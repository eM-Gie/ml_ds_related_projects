```python
import pandas as pd
food = pd.read_csv("food.txt",delimiter=" ",comment="#", header=None)
food
```

```python
X = np.array(food)[:,1:].T.astype("float") # Wiersze to kraje
print(X.shape)
y = food.columns[1:]
print(y)
```

```python
pd.DataFrame(X)
```

```python
df=pd.DataFrame(X)
sns.pairplot(df)
plt.show()
print("sdf")
```

```python
from sklearn.preprocessing import StandardScaler
df
X_std = StandardScaler().fit_transform(X)
```

```python
from sklearn.decomposition import PCA
model = PCA()
pca_features=model.fit_transform(X_std)

features = range(model.n_components_)

plt.bar(features,model.explained_variance_ratio_)
plt.xticks(features)

plt.show()
```
