# Wielowymiarowy rozkład normalny

```python
import matplotlib.pylab as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.mlab as mlab
from matplotlib import cm
from scipy import stats
```

# Zad. 
Narysujmy gęstość dwuwymiarowego rozkładu Normalnego o parametach 

* n=[0,0], cov = [[1,0],[0,1]]
* n=[0,0], cov = [[2,0],[0,1]]
* n=[0,0], cov = [[1,0],[0,2]]

```python
from scipy.stats import multivariate_normal
from mpl_toolkits.mplot3d import Axes3D


def density_normal_distribution(n,cov):
    
    x = np.linspace(-10,10,500)
    y = np.linspace(-10,10,500)
    X, Y = np.meshgrid(x,y)
    pos = np.empty(X.shape + (2,))
    pos[:, :, 0] = X; pos[:, :, 1] = Y


    rv = multivariate_normal(n, cov)

    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')
    ax.plot_surface(X, Y, rv.pdf(pos),cmap='viridis')
    plt.show()
    

    
density_normal_distribution([0,0],[[1,0],[0,1]])
density_normal_distribution([0,0],[[2,0],[0,1]])
density_normal_distribution([0,0],[[2,0],[0,2]])

```

# Zad 
Narysujmy kontury gęstości dwuwymiarowego rozkładu Normalnego o parametach 

* n=[0,0], cov = [[1,0],[0,1]]
* n=[0,0], cov = [[2,0],[0,1]]
* n=[0,0], cov = [[1,0],[0,2]]

```python
import seaborn as sns


def density_contour(n,cov):
    x, y = np.mgrid[-5:5:.1, -5:5:.1]
    rv = multivariate_normal(n,cov)
    data = np.dstack((x, y))
    z = rv.pdf(data)
    plt.contourf(x, y, z)
    plt.show()


density_contour([0,0],[[1,0],[0,1]])
density_contour([0,0],[[2,0],[0,1]])
density_contour([0,0],[[2,0],[0,2]])

```

# Zad 
Wylosujmy próbkę z dwuwymiarowego rozkładu Normalnego o parametach 

* n=[0,0], cov = [[1,0],[0,1]]
* n=[0,0], cov = [[2,0],[0,1]]
* n=[0,0], cov = [[1,0],[0,2]]

i narysuj odpowiedni wykres

```python

def scattered(n,cov):
    X1_rv=multivariate_normal(n, cov)
    X = X1_rv.rvs(1000)
    fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(5, 5))
    ax.scatter(X[:, 0], X[:, 1])
    ax.set_xlim([-5,5])
    ax.set_ylim([-5,5])
    plt.show()

scattered([0,0],[[1,0],[0,1]])
scattered([0,0],[[2,0],[0,1]])
scattered([0,0],[[2,0],[0,2]])
```
