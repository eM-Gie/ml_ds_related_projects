# Underfitting vs. Overfitting

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn import  metrics
```

Rozważmy następujący zbiór punktów:

```python
np.random.seed(0)
n_samples = 30
true_fun = lambda X: np.cos(1.5 * np.pi * X)
X = np.sort(np.random.rand(n_samples))
y = true_fun(X) + np.random.randn(n_samples) * 0.1
```

```python
# Plot outputs
plt.figure(figsize=(6,6));
axes = plt.gca()
axes.set_xlim([0,1])
axes.set_ylim([-1.5,1.5])
plt.scatter(X, y,  color='black')
x_tr = np.linspace(0, 1, 200)
plt.show()
```

Podziel dane na dwie grupy.

```python
s=np.random.random_sample(n_samples)
s[s>0.5]=1
s[s<=0.5]=0
s
X1=X[s==1]
y1=y[s==1]
X2=X[s==0]
y2=y[s==0]

# Plot outputs
plt.figure(figsize=(6,6));
axes = plt.gca()
axes.set_xlim([0,1])
axes.set_ylim([-1.5,1.5])
plt.scatter(X1, y1,  color='blue')
plt.scatter(X2, y2,  color='red')
x_tr = np.linspace(0, 1, 200)
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn import  metrics
```

```python
from sklearn import  metrics
s=np.random.random_sample(n_samples)
s[s>0.5]=1
s[s<=0.5]=0
s
X1=X[s==1]
y1=y[s==1]
X2=X[s==0]
y2=y[s==0]

 


model1 = make_pipeline(PolynomialFeatures(1), LinearRegression() )
model1.fit(np.vstack(X2), y2)
print(metrics.r2_score(y1, model1.predict(np.vstack(X1))))

 


model2 = make_pipeline(PolynomialFeatures(1), LinearRegression() )
model2.fit(np.vstack(X1), y1)
print(metrics.r2_score(y2, model2.predict(np.vstack(X2))))

 

# Plot outputs
x_plot = np.vstack(np.linspace(0, 10, 300))
plt.figure(figsize=(6,6));
axes = plt.gca()
axes.set_xlim([0,1])
axes.set_ylim([-1.5,1.5])
plt.scatter(X1, y1,  color='blue')
plt.plot(x_plot, model1.predict(x_plot), 'r');
plt.plot(x_plot, model2.predict(x_plot), 'b');

 

plt.scatter(X2, y2,  color='red')
x_tr = np.linspace(0, 1, 200)
plt.show()
```

```python
from sklearn import  metrics
s=np.random.random_sample(n_samples)
s[s>0.5]=1
s[s<=0.5]=0
s
X1=X[s==1]
y1=y[s==1]
X2=X[s==0]
y2=y[s==0]

 


model1 = make_pipeline(PolynomialFeatures(20), LinearRegression() )
model1.fit(np.vstack(X2), y2)
print(metrics.r2_score(y1, model1.predict(np.vstack(X1))))

 


model2 = make_pipeline(PolynomialFeatures(20), LinearRegression() )
model2.fit(np.vstack(X1), y1)
print(metrics.r2_score(y2, model2.predict(np.vstack(X2))))

 

# Plot outputs
x_plot = np.vstack(np.linspace(0, 10, 300))
plt.figure(figsize=(6,6));
axes = plt.gca()
axes.set_xlim([0,1])
axes.set_ylim([-1.5,1.5])
plt.scatter(X1, y1,  color='blue')
plt.plot(x_plot, model1.predict(x_plot), 'r');
plt.plot(x_plot, model2.predict(x_plot), 'b');

 

plt.scatter(X2, y2,  color='red')
x_tr = np.linspace(0, 1, 200)
plt.show()
```
