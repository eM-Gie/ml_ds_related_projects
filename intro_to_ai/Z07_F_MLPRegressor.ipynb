# MLPRegressor

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import sklearn.linear_model as lm
from sklearn import datasets, linear_model
import matplotlib.pyplot as plt

from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import GridSearchCV
```

Proszę narysować wykres dowolnej funkcji na odcinku $[0,3]$.

Będzie to model, który będziemy chcieli znaleźć.

```python
f = lambda x: ((x*2-1)*(x**2-2)*(x-2)+3)
x_tr = np.linspace(0, 3, 200)
y_tr = f(x_tr)
x = stats.uniform(0,3).rvs(100)
y = f(x) + stats.norm(0,0.2).rvs(len(x))
```

# Zadanie
Wykonaj regresję z różnymi parametrami.

```python
x=np.vstack(x)
x_plot = np.vstack(np.linspace(0, 10, 100))

MLP = MLPRegressor(hidden_layer_sizes=(12, 100,10), activation='tanh', max_iter=5000)
y_rbf = MLP.fit(x,y)

# Plot outputs
plt.figure(figsize=(6,6));
axes = plt.gca()
axes.set_xlim([0,3])
axes.set_ylim([0,8])
plt.scatter(x, y,  color='black')
plt.plot(x_plot, MLP.predict(x_plot), color='blue',linewidth=3)
plt.show()

```

```python
from sklearn import  metrics

metrics.r2_score(y, MLP.predict(x))
```

Wyznacz optymalne parametry dla modelu MLPRegressin.

```python
import pandas as pd
df_adv = pd.read_csv('Advertising.csv', index_col=0)
X = df_adv[['TV', 'Radio','Newspaper']]
y = df_adv['Sales']
df_adv.head()

grid= GridSearchCV(MLPRegressor(),
        cv=5,
        param_grid={"activation": ['identity', 'logistic', 'tanh', 'relu'],
                    "alpha": [0.00001,0.0001,0.001,0.1,1,10,100,1000],
                    "hidden_layer_sizes": [(1,),(10,),(100,),(1000,)]})
 
grid.fit(X,y)
grid.best_params_
```
